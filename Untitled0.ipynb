```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Road Accident Data.csv")
print(df.info())
print(df.head())
```

```python
df=df.dropna()
df=df.drop('Accident_Index',axis=1)
df=df.drop('Accident Date',axis=1)
```

```python
correlation=df.corr(numeric_only=True)
print(correlation)
```

```python
plt.hist(df['Number_of_Casualties'],bins=10,color='green',edgecolor='black')
plt.title('Distribution of casualties (histogram)')
plt.xlabel('Number of casualties')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.boxplot(df['Number_of_Casualties'])
plt.title("Boxplot of Number of Casualties")
plt.show()
```

```python
x=df['Accident_Severity']
y=df['Road_Type']
plt.bar(x,y,color='brown')
plt.title('Accident Severity based on type of road (barplot)')
plt.xlabel('Type of Road')
plt.ylabel('Severity of Accident')
plt.show()
```

```python
x=df['Number_of_Vehicles']
y=df['Speed_limit']
plt.scatter(x,y,color='blue')
plt.title('Number_of_Vehicless vs Speed_Limit (scatterplot)')
plt.xlabel('Number of Vehicles')
plt.ylabel('Speed Limit')
plt.show()
```

```python
df=df.head()
y=df['Speed_limit']
lab=df['Number_of_Casualties']
plt.pie(y,labels=lab)
plt.legend()
plt.show()
```
